```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
movies_df = pd.read_csv('C:/Users/USUARIO/Desktop/SOYhENRY/sistema de recomendacion/Movies/CSV/movies_datasetc.csv') 

# Vista inicial de las columnas relevantes
relevant_columns = ['title', 'release_date', 'popularity']
print("Columnas relevantes:")
print(movies_df[relevant_columns].head())

# 1. **Verificar valores nulos**
print("\nValores nulos por columna:")
print(movies_df[relevant_columns].isnull().sum())
```

```python

# 2. **Distribución de los puntajes de popularidad**
plt.figure(figsize=(10, 6))
sns.histplot(movies_df['popularity'], bins=30, kde=True, color='blue')
plt.title('Distribución de la popularidad de las películas')
plt.xlabel('Puntaje de Popularidad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# 3. **Estadísticas descriptivas de popularidad**
print("\nEstadísticas descriptivas de la popularidad:")
print(movies_df['popularity'].describe())
```

```python
# Eliminar nulos en 'release_date'
movies_df = movies_df.dropna(subset=['release_date'])

# Convertir 'release_date' a formato datetime y extraer el año
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
movies_df['release_year'] = movies_df['release_date'].dt.year

# Eliminar filas con fechas inválidas (NaT)
movies_df = movies_df.dropna(subset=['release_year'])

# Agrupar por año de lanzamiento y calcular la popularidad promedio
popularity_by_year = movies_df.groupby('release_year')['popularity'].mean()

# Visualización
plt.figure(figsize=(12, 6))
popularity_by_year.plot(kind='line', color='green')
plt.title('Popularidad promedio por año de estreno')
plt.xlabel('Año de Estreno')
plt.ylabel('Popularidad Promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 5. **Verificar títulos duplicados**
duplicated_titles = movies_df['title'].duplicated().sum()
print(f"\nCantidad de títulos duplicados: {duplicated_titles}")

# Si hay duplicados, mostrarlos para análisis
if duplicated_titles > 0:
    print("\nEjemplo de títulos duplicados:")
    print(movies_df[movies_df['title'].duplicated(keep=False)][['title', 'release_date']].head(10))

```

```python
# 6. **Identificar las películas con mayor y menor popularidad**
most_popular = movies_df.loc[movies_df['popularity'].idxmax()]
least_popular = movies_df.loc[movies_df['popularity'].idxmin()]

print("\nPelícula con mayor popularidad:")
print(most_popular[['title', 'release_date', 'popularity']])

print("\nPelícula con menor popularidad:")
print(least_popular[['title', 'release_date', 'popularity']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
movies_df = pd.read_csv('C:/Users/USUARIO/Desktop/SOYhENRY/sistema de recomendacion/Movies/CSV/movies_datasetc.csv') 

# Vista inicial de las columnas relevantes
relevant_columns = ['title', 'release_date', 'popularity', 'budget', 'revenue']
print("Columnas relevantes:")
print(movies_df[relevant_columns].head())

# 1. **Verificar valores nulos**
print("\nValores nulos por columna:")
print(movies_df[relevant_columns].isnull().sum())

# Conversión de release_date a fecha y extracción del año
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
movies_df['release_year'] = movies_df['release_date'].dt.year

# Eliminar filas con fechas no válidas
movies_df = movies_df.dropna(subset=['release_year'])

# 2. **Distribución de la popularidad**
plt.figure(figsize=(10, 6))
sns.histplot(movies_df['popularity'], bins=30, kde=True, color='blue')
plt.title('Distribución de la popularidad de las películas')
plt.xlabel('Puntaje de Popularidad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. **Estadísticas descriptivas de popularidad**
print("\nEstadísticas descriptivas de la popularidad:")
print(movies_df['popularity'].describe())

# 4. **Correlación entre variables numéricas**
correlation_matrix = movies_df[['popularity', 'budget', 'revenue']].corr()
print("\nMatriz de correlación:")
print(correlation_matrix)

# Visualización de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()

# 5. **Valores atípicos (outliers) en popularidad**
plt.figure(figsize=(10, 6))
sns.boxplot(data=movies_df, x='popularity', color='purple')
plt.title('Valores atípicos en popularidad')
plt.tight_layout()
plt.show()

```

```python
# Convertir la columna 'popularity' a valores numéricos, reemplazando errores por NaN
movies_df['popularity'] = pd.to_numeric(movies_df['popularity'], errors='coerce')

# 6. Tendencias temporales: popularidad promedio por año
popularity_by_year = movies_df.groupby('release_year')['popularity'].mean()

plt.figure(figsize=(12, 6))
popularity_by_year.plot(kind='line', color='green')
plt.title('Popularidad promedio por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Popularidad promedio')
plt.grid(True)
plt.show()
```

```python
# 7. **Relación entre popularidad, presupuesto y revenue**
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies_df, x='budget', y='popularity', hue='revenue', size='revenue', palette='viridis')
plt.title('Relación entre Presupuesto, Popularidad y Revenue')
plt.xlabel('Presupuesto')
plt.ylabel('Popularidad')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 8. **Películas más populares por década**
movies_df['decade'] = (movies_df['release_year'] // 10) * 10
popular_by_decade = movies_df.groupby('decade')['popularity'].idxmax()
most_popular_movies = movies_df.loc[popular_by_decade, ['title', 'decade', 'popularity']]
print("\nPelículas más populares por década:")
print(most_popular_movies)
```

```python
# 9. **Identificar títulos duplicados**
duplicated_titles = movies_df['title'].duplicated().sum()
print(f"\nCantidad de títulos duplicados: {duplicated_titles}")
if duplicated_titles > 0:
    print("\nEjemplo de títulos duplicados:")
    print(movies_df[movies_df['title'].duplicated(keep=False)][['title', 'release_year']].head(10))
```

```python
# 10. **Película con mayor y menor popularidad**
most_popular = movies_df.loc[movies_df['popularity'].idxmax()]
least_popular = movies_df.loc[movies_df['popularity'].idxmin()]
print("\nPelícula con mayor popularidad:")
print(most_popular[['title', 'release_year', 'popularity']])
print("\nPelícula con menor popularidad:")
print(least_popular[['title', 'release_year', 'popularity']])
```
